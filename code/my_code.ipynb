### Make my own Holdout Set I can score later

```python
houses_holdout = houses.loc[1501:,:]
houses = houses.loc[0:1500,:]
```

```python
print(houses.shape)
print(houses_holdout.shape)
```

### Set up features, X and y

```python
features = ['Lot Area', 'Bedroom AbvGr']

X = houses[features]
y = houses['SalePrice']
```

### Continue down two paths, concurrently:
- The 1st path: `.fit_transform(X)`, then test-train-split
- The 2nd path: Test-train-split, then `.fit_transform(X_train)` and `.transform(X_test)`

```python
ss_pre = StandardScaler()
X_prescaled = ss_pre.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train_prescaled, X_test_prescaled, y_train_prescaled, y_test_prescaled = train_test_split(X_prescaled, y)
```

```python
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.fit(X_test)
```

### Instantiate all modeling objects, and check cross_val_score

```python
linreg = LinearRegression()
ridge = Ridge()
lasso = Lasso()

linreg_prescaled = LinearRegression()
ridge_prescaled = Ridge()
lasso_prescaled = Lasso()
```

```python
linreg_scores = cross_val_score(linreg, X_train_scaled, y_train)
ridge_scores = cross_val_score(ridge, X_train_scaled, y_train)
lasso_scores = cross_val_score(lasso , X_train_scaled, y_train)

linreg_pre_scores = cross_val_score(linreg_prescaled, X_train_prescaled, y_train_prescaled)
ridge_pre_scores = cross_val_score(ridge_prescaled, X_train_prescaled, y_train_prescaled)
lasso_pre_scores = cross_val_score(lasso_prescaled, X_train_prescaled, y_train_prescaled)
```

```python
print(np.mean(linreg_scores))
print(np.mean(ridge_scores))
print(np.mean(lasso_scores))
print()
print(np.mean(linreg_pre_scores))
print(np.mean(ridge_pre_scores))
print(np.mean(lasso_pre_scores))
```

- The "pre-scaled" appears to score better at the cross-val stage. But will this hold true on "unseen" data?

### Score all models against "unseen" data, (my Holdout set)

```python
X_holdout = houses_holdout[features]

# My post-split scaler
X_holdout_scaled = ss.transform(X_holdout)
# My pre-split scaler
X_holdout_prescaled = ss_pre.transform(X_holdout)

y_holdout = houses_holdout['SalePrice']
```

```python
linreg.fit(X_train_scaled, y_train)
ridge.fit(X_train_scaled, y_train)
lasso.fit(X_train_scaled, y_train)

linreg_prescaled.fit(X_train_prescaled, y_train_prescaled)
ridge_prescaled.fit(X_train_prescaled, y_train_prescaled)
lasso_prescaled.fit(X_train_prescaled, y_train_prescaled)
```

```python
print(linreg.score(X_holdout_scaled, y_holdout))
print(ridge.score(X_holdout_scaled, y_holdout))
print(lasso.score(X_holdout_scaled, y_holdout))

print(linreg_prescaled.score(X_holdout_prescaled, y_holdout))
print(ridge_prescaled.score(X_holdout_prescaled, y_holdout))
print(lasso_prescaled.score(X_holdout_prescaled, y_holdout))
```
